```python
import csv
import json

def convert_csv_to_ts():
    # 读取CSV文件
    personnel_data = []
    
    with open('sme.csv', 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            # 确定location
            location = ""
            if "金山" in row['部门']:
                location = "金山"
            elif "徐汇" in row['部门']:
                location = "徐汇"
            elif "银川" in row['部门']:
                location = "银川"
            elif "沈阳" in row['部门']:
                location = "沈阳"
            else:
                location = "未知"  # 默认金山
            
            # 构建人员数据
            person = {
                "name": row['姓名'],
                "email": row['邮箱'],
                "phone": "",
                "department": row['部门'],
                "major": row['专业'],
                "projects": [p.strip() for p in row['项目经历'].split('、') if p.strip()],
                "industries": [i.strip() for i in row['擅长领域'].split('、') if i.strip()],
                "skills": [s.strip() for s in row['技能'].split('、') if s.strip()],
                "qualifications": [q.strip() for q in row['资质'].split('、') if q.strip()],
                "regions": [],
                "location": location
            }
            personnel_data.append(person)
    
    # 生成TypeScript文件内容
    ts_content = """export const personnelData = """ + json.dumps(personnel_data, ensure_ascii=False, indent=2)
    
    # 写入TypeScript文件
    with open('SME_Person_data.tsx', 'w', encoding='utf-8') as f:
        f.write(ts_content)

if __name__ == "__main__":
    convert_csv_to_ts()
```
